```python
# 1. 고객의 정보는 이름, 계좌번호, 예금금액만 관리할 것(0)
# 2. 계좌번호는 정수값을 가지되, 중복값을 허용하지 않는다.(0)
# 3. 둘 이상의 고객에 대한 정보를 저장할 수 있도록 구현할 것.(0)
# 4. 계좌개설, 입금처리, 출금처리, 전체고객에 대한 계좌잔액 출력에 관한 로직을 구현할 것.(세모)
# 5. 입금처리는 계좌번호를 입력받아 해당계좌의 존재유무에 따라 처리가 되도록 구현하되,
#    입력한 계좌번호가 존재하지 않을 경우 "계좌번호가 존재하지 않습니다." 라는 메시지가 출력되도록
#    구현할 것.(0)
# 6. 출금처리는 계좌번호를 입력받아 해당계좌의 존재유무에 따라 처리하는 로직은 입금처리와 동일하며,
#    출금처리시 계좌의 잔액 유무에 따른 메시지가 출력되도록 구현할 것.
#    즉, 출금금액보다 예금잔액이 부족할 경우 "잔액이 부족합니다."라는 메시지 출력할 것.
# 7. 기타 처리 기능은 일반적으로 통용되는 기능을 이용할 것.
# 8. 아래 출력형식에 맞게 프로그램을 구현할 것.
```

```python
customer_list = {}

num = "111"
name = "aaa"
cash = 10000
customer = Bank(num, name, cash)
customer_list[num] = customer

print(customer_list[num].All())
```

```python
## 프로그램
#중복방지를 위해 딕셔너리 개설
customerList = dict()
#계좌개설을 위한 메서드
def openBank():
    num = input('계좌번호 =')
    
    check = 1 ## 이거 1로 바꿨음
    while num in customerList.keys():
        print('이미 사용되고 있는 계좌번호입니다.')
        print('다시 사용하고자 하는 계좌번호를 입력해주세요.')
        num = input('계좌번호 =')
        
        if check == 3:
            
            print('3번 잘못입력하셨습니다.')
            return
        
        check += 1
             
            
        
    name = input('고객이름 =')
    cash = int(input('예금금액 ='))
    customer = Bank(num, name, cash) 
    customerList[num] = customer
    print(name + ' 고객님 환영합니다')
    print(name + ' 고객님의 계좌 개설이 완료되었습니다.')

        

#입금처리를 위한 메서드

def inPut(): 
    print('== 입금처리 ==')
    num = input("계좌번호 = ")
    while num in customerList.keys():
        print(customerList[num].getName(),'고객님 반갑습니다.')
        print('입금금액을 입력해주세요.')
        money = int(input('입금금액 = '))
        customerList[num].inmoney(money)
        print('-'*10)
        print('입금이 처리되었습니다.')
        print('현재 계좌의 총 예금금액은',customerList[num].elsemoney(),'원 입니다.')
        break
    else:
        print('-'*10)
        print('계좌번호가 존재하지 않습니다.')
        
        
            
    
#출금처리를 위한 메서드
def out():
    print('== 출금처리 ==')
    num = input("계좌번호 = ")
    while num in customerList.keys():
        print(customerList[num].getName(),'고객님 반갑습니다.')
        print('출금금액을 입력해주세요.')
        money = int(input('출금금액 = '))
        if customerList[num].outmoney(money) > 0:   
            print('출금이 처리되었습니다.')
            print('현재 계좌의 총 예금금액은',customerList[num].elsemoney(),'원 입니다.')
    
        else:
            print('고객님의 계좌에 잔액이 부족하여')
            print('고객님께서 입력하신',money,'원은 출금할 수 없습니다.')
        
        break
        
    else:
        print('입력하신 계좌번호는 없는 계좌번호입니다.')
        print('다시 사용하고자 하는 계좌번호를 입력해주세요.')   

    
#전체고객 조회를 위한 메서드
def alldisp():
    print( '== 전체고객 잔액조회 ==')
    print('계좌번호','고객이름','예금잔액', sep = "\t")
    print('-'*20)
    for i in customerList.keys():
        customerList[i].All()
```

```python
# 메뉴와 메뉴 출력하는 메서드
print('은행프로그램을 실행합니다.')
while True:
    print('== 계좌관리 프로그램 ==')
    print('1. 계좌개설')
    print('2. 입금처리')
    print('3. 출금처리')
    print('4. 전체고객 잔액조회')
    print('5. 프로그램 종료')
    print('-'*20)
    select = int(input('메뉴 = '))
    if select == 1:
        openBank()
    elif select == 2:
        inPut()
    elif select == 3:
        out()
    elif select == 4:
        alldisp()
    elif select == 5:
         break
    else:
        print('잘못된 선택입니다.')
        
print('프로그램을 종료합니다.')
```

```python
## 잘못됨
# Entity Class : 단순히 데이터를 담아서 전달할 목적으로 사용하는 클래스
# 한 명의 직원에 대한 정보를 관리할 목적의 클래스
class Customer:
    
    def __init__(self, name, num, pay): # 이름, 계좌번호, 예금금액, 입금금액, 출금금액
        self.name = name
        self.num = num
        self.pay = pay
    

    def getName(self):
        return self.name
    
    def setName(self, name):
        self.name = name
    
    def getNum(self):
        return self.num
    
    def setNum(self, num):
        self.num = num
    
    def getPay(self):
        return self.pay
    
    def setPay(self, pay):
        self.pay = pay
        
    
    def disp(self):
        return "{0}         {1}          {2}".format(self.getNum(), self.getName(), self.getPay())
    
```

```python
# Handler Class = Controller 
# 실제 처리를 담당하는 클래스, 프로그램 전체의 흐름을 관리하는 클래스

class C_controll:
    
    def __init__(self):
        self.custList = [] 
        
    def addCust(self): #계좌개설
        print('== 계좌개설 ==')
        num = int(input("계좌번호 = "))
        name = input("고객이름 = ")
        pay = int(input("예금금액 = "))

        cus = Customer(num, name, pay)
        self.custList.append(cus)
        print('-'*10)
        print("계좌개설이 완료되었습니다.")
        
    def allCust(self): #고객전체조회
        if len(self.custList) != 0:
            for cus in self.custList:
                print('== 전체고객 잔액조회 ==')
                print('계좌번호   고객이름   예금잔액')
                print(cus.disp())
                print('총 보유예금액 :', cus.getPay())
            print("고객정보 조회가 완료되었습니다.")
        else:
            print("등록된 고객의 정보가 존재하지 않습니다.")
            
    def oneCust(self): #입금
        print('== 입금처리 ==')
        num = input("계좌번호 = ")
        check = 0 

        for cus in self.custList:   # range(len(empList))
            if num == cus.getNum():
                check = 1    
                money = int(input('입금금액:'))
                if money > 0: 
                    print('-'*10)
                    print(money + pay)     
                
        if check == 0:
            print('-'*10)
            print("해당 계좌번호가 존재하지 않습니다.") 
            
            
    def twoCust(self): #입금
        print('== 출금처리 ==')
        num = input("계좌번호 = ")
        check = 0 

        for cus in self.custList:   # range(len(empList))
            if num == cus.getNum():
                check = 1    
                coin = int(input('출금금액:'))
                if pay - coin > 0: 
                    print('-'*10)
                    print(pay - coin) 
                print('잔액이 부족합니다.')
                
        if check == 0:
            print('-'*10)
            print("해당 계좌번호가 존재하지 않습니다.") 
   
    def showMenu(self): #메뉴
        print("\n == 계좌관리 프로그램 ==")
        print("1. 계좌개설")
        print("2. 입금처리")
        print("3. 출금처리")
        print("4. 전체고객 잔액조회")
        print("5. 프로그램 종료")
```

```python
#### 프로그램을 동작하는 역할을 수행하는 함수
def main():
    
    handle = C_controll()
    
    while True:
        handle.showMenu()
        menu = input("메뉴 = ")
        
        if menu == "1":
            handle.addCust()
        elif menu == "2":
            handle.oneCust() 
        elif menu == "3":
            handle.twoCust()
        elif menu == "4":
            handle.allCust()
        elif menu == "5":
            print("프로그램을 종료합니다.")
            break
        else:
            print("메뉴의 선택이 올바르지 않습니다.")
            
main()
```
