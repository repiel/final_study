```python
# 수치 데이터의 기초통계량
chipo.describe().T
```

```python
# 테이블 번호로 그룹
# 총 주문한 테이블의 수와 테이블별로 주문한 음식 수량을 알 수 있다.
chipo.groupby('order_id').count()
```

```python
# 총 주문 테이블 => 1834 테이블인 것을 알 수 있다.
chipo.groupby('order_id').size()
```

```python
# 음식의 종류 및 각 음식의 매출 개수
# 가장 매출이 많은 음식 => Chicken Bowl인 것을 알 수 있다.
# 따라서 가장 인기가 많은 음식도 Chicken Bowl이다.
chipo.groupby(['item_name']).size()
```

```python
# 가게에서 가장 가격이 높은 음식 구하기
chipo['item_price'].sort_values(ascending=False)
```

```python
# 메뉴의 개별 가격
# Bowl 종류의 선택메뉴(choice_description)가 포함되지 않은 기본가격이 가장 높은 것을 알 수 있다.
chipo.sort_values(by = 'item_price')
```

```python
# 테이블에서 주문시켰을 때 가장 가격 매출이 높은 메뉴(선택메뉴(choice_description) 포함)
chipo.sort_values(by = ['order_id','item_price'])
```

```python
# chipo의 item_price를 Numeric 타입으로 바꾸기
chipo['item_price'] = chipo['item_price'].str.lstrip('$')
chipo['item_price'] = chipo['item_price'].astype(float)
```

```python
# 테이블 별 매출 금액을 알 수 있다. 
# 최대와 최소 테이블 별 매출 금액을 알 수 있다.
chipo.groupby(['order_id'])[['item_price']].sum()
```
