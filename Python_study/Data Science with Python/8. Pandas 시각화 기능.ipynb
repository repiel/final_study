- pandas의 시리즈나 데이터프레임에 plot이라는 메서드가 내장되어 있다.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
# matplotlib 한글화
mpl.rcParams['font.family']='D2coding' # 원하는 글꼴
```

```python
np.random.seed(0)

df= pd.DataFrame(np.random.randn(100, 3),
                index = pd.date_range('2018-1-1', periods = 100),
                columns = list('ABC')).cumsum()
df
```

```python
df.plot()
plt.title('Pandas의 내장된 plot 메서드 사용 예')
plt.xlabel("날짜")
plt.ylabel('Data')
plt.show()
```

- plot 메서드에 kind 인수를 통해 다양한 플롯을 시각화 할 수 있다.
- bar, pie, hist, kde, box, scatter

```python
import seaborn as sns

iris = sns.load_dataset('iris')
titanic = sns.load_dataset('titanic')
```

```python
iris.sepal_length[:20].plot(kind='bar')
plt.title('iris의 꽃받침의 길이를 이용한 시각화')
plt.xlabel('data')
plt.ylabel('꽃받침의 길이')
plt.show()
```

```python
iris[:5].plot.bar()
plt.title('Iris 붓꽃 데이터의 막대 그래프')
plt.xlabel('Data')
plt.ylabel('각 Feature의 값')
plt.ylim(0, 7)
plt.legend(loc=1)
plt.show
```

```python
# 품종별로 각 특성의 평균값을 시각화하시오.
df2 = iris.groupby(iris.species).mean()
df2
```

```python
df2.plot.bar()
plt.title('품종별 Feature의 평균 막대 그래프')
plt.xlabel('품종')
plt.ylabel('평균')
plt.ylim(0, 7)
plt.legend(loc=0)
plt.show
```

```python
df2.T.plot.bar()
plt.title('품종별 Feature의 평균 막대 그래프')
plt.xlabel('품종')
plt.ylabel('평균')
plt.ylim(0, 7)
plt.legend(loc=0)
plt.show
```

```python
# 객실등급별 승객수의 비율을 시각화
df3 = titanic.pclass.value_counts()
df3
```

```python
df3.plot.pie(autopct='%1.2f%%')
plt.axis('equal')
plt.show()
```

```python
iris.plot.kde(cmap="crest")
plt.show()
```

```python
iris.plot.hist(cmap="crest")
plt.show()
```

```python
iris.plot.hist(cmap=sns.cubehelix_palette(as_cmap=True))
plt.show()
```

```python
iris.plot.box(cmap=sns.cubehelix_palette(as_cmap=True))
plt.show()
```

```python
iris.boxplot(by='species')
plt.tight_layout(pad=1, h_pad = 2)
plt.show()
```
