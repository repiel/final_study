# Groupby(그룹별 집계)
- groupby 매커니즘 : 분할-적용-결합

```python
ipl_data = {'Team':['Riders','Riders','Devils','Devils','Kings','Kings','Kings','Kings',
                    'Riders','Royals','Royals','Riders'],
           'Rank':[1,2,2,3,3,4,1,1,2,4,1,2],
           'Year':[2014,2015,2014,2015,2014,2015,2016,2017,2016,2014,2015,2017],
           'Points':[876,789,863,673,741,812,756,788,694,701,804,690]}
df = pd.DataFrame(ipl_data)
df
```

- Series의 데이터 그룹화: 특정열인덱싱.groupby(기준열인덱싱)
- DataFrame의 데이터 그룹화: df.groupby('기준이 될 컬럼명')

```python
# Team별 points의 합을 추출
df.groupby('Team')['Points'].sum()
```

```python
# Team과 Year별로 points의 합을 추출
multi_groupby = df.groupby(['Team','Year'])['Points'].sum()
multi_groupby
```

```python
multi_groupby.index
```

```python
multi_groupby['Devils':'Kings']
```

```python
# swaplevel : 인덱스 간의 레벨 변경
multi_groupby.swaplevel().sort_index()
```

```python
multi_groupby.sum(level=0)
```

```python
multi_groupby.sum(level=1)
```

- 집계(aggreate) : 그룹화된 통계 정보를 추출할 때 사용하는 함수, agg
- 변환(transform)
- 필터(filter) : 특정 정보를 제거하여 보여주는 필터링 기능

```python
grouped = df.groupby('Team')
grouped
```

```python
# get_group() : 키 값을 기준으로 분할된 데이터프레임 객체를 반환
grouped.get_group('Riders')
```

```python
# agg() : 다양한 함수를 한번에 적용하는 함수
grouped.agg(min)
```

```python
grouped.agg(np.mean)
```

```python
grouped.agg([sum, mean])
```

```python
grouped.transform(max)
```

```python
z_value = lambda x : (x - x.mean()) / x.std()
grouped.transform(z_value)
```

```python
# 필터(Filter) : 특정 조건으로 데이터를 검색할 때 사용하는 함수
df.groupby('Team').filter(lambda x : len(x) >= 3)
```

```python
# 우선 최대값 확인
df.groupby('Team')['Points'].max()
```

```python
# 팀별로 묶고 points의 최대값이 850을 초과하는 자료만 추출
df.groupby('Team').filter(lambda x : x['Points'].max() > 850)
```

### tips datasets 

```python
import seaborn as sns
tips = sns.load_dataset('tips')
```

```python
tips.info()
```

```python
tips.tail()
```

```python
# 어떤 경우에 팁을 더 많이 줄까?
# 가장 팁을 많이 주는 상황 찾기
# 청구서금액 대비 팁의 비율 
tips['tip_pct'] = tips['tip'] / tips['total_bill']
tips.head()
```

```python
# 수치 데이터의 기초통계량
tips.describe()
```

```python
# 성별로 그룹
tips.groupby('sex').count()
```

```python
tips.groupby('sex').size()
```

```python
# 성별과 흡연유무로 나누어 데이터의 갯수 분석
tips.groupby(['sex', 'smoker']).size()
```

```python
# 성별에 따른 평균 팁 비율
tips.groupby('sex')['tip_pct'].mean()
```

```python
# 흡연과 성별에 따른 평균 팁 비율
tips.groupby(['sex','smoker'])['tip_pct'].mean()
```

```python
tips.groupby(['sex','smoker'])['tip_pct'].max() - tips.groupby(['sex','smoker'])['tip_pct'].min()
```

```python
tips.pivot_table('tip_pct', 'sex', 'smoker', aggfunc = 'count', margins = True)
```

```python
tips.groupby(['sex','smoker'])[['tip_pct']].mean()
```

```python
tips.pivot_table('tip_pct', 'sex', 'smoker')
```

```python
tips.groupby(['sex','smoker'])[['tip_pct']].describe()
```

```python
# 각 그룹에서 가장 많은 팁과 가장 적은 팁의 차이를 확인하시오.
tips.groupby(['sex','smoker'])[['tip_pct']]
```

```python
def max_min(x):
    return x.max() - x.min()

tips.groupby(['sex','smoker'])[['tip']].agg(max_min)
```

```python
tips.groupby(['sex','smoker'])[['tip']].agg(['mean', max_min])
```

```python
tips.groupby(['sex','smoker'])[['tip']].agg([np.mean, max_min])
```

```python
tips.groupby(['sex','smoker']).agg({'tip':'mean', 'total_bill': max_min})
```
