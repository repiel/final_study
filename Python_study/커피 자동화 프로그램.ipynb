```python
# 커피 자동 판매기 프로그램을 구현하시오.
# 커피 판매기에는 커피종류별 재료가 아래와 같이 셋팅되어있다.
# 커피 3, 물 3, 설탕 3, 프림 3, 컵 3
# 블랙커피 = 커피1 + 물1 + 컵1
# 설탕커피 = 커피1 + 물1 + 설탕1 + 컵1
# 프림커피 = 커피1 + 물1 + 프림1 + 컵1
# 밀크커피 = 커피1 + 물1 + 설탕1 + 프림1 + 컵1
# 만약 재료가 소진되면 해당 커피의 제조가 불가하여 판매가 불가능하다.
# 각 재료의 보충시 모든 재료가 3으로 셋팅되거나, 필요한 재료만 3으로 셋팅할 수 있다.
# 각 재료의 최대 셋팅 개수는 3이다.
# 커피가격은 블랙커피(1000), 설탕커피(1000), 프림커피(1500), 밀크커피(2000)

# ================================
# ======  커피 자동 판매기 동작  ======
# ================================
# 1. 블랙커피(1000원)
# 2. 설탕커피(1000원)
# 3. 프림커피(1500원)
# 4. 밀크커피(2000원)
# --------------------------------
# 투입금액 =
# 주문번호 = 

# 투입금액 = manager  **이스터 에그**
# ==== 자동판매기 관리 프로그램 ====
# [1]재료보충 [2]금액조정 [3]작업종료
# ============================
# 메뉴 = 1
 
#         ===== 재료보충 ===== 
# [1]전체 [2]커피 [3]물 [4]설탕 [5]프림 [6]컵
# ======================================
# 선택 = 


###문제점들
# 돈 받는게 안된다..
# 매니저의 금액 변경이 안됨...
# 커피추출은 가능!
# 재료가  -되어서 커피추출이 되지 않아야 하는 기능도 넣어야 함
```

```python
class Cafe:
    
    # 커피종류
    def __init__(self, black_coffe, suger_coffe, prim_coffe, milk_coffe, money): 
        self.black_coffe = black_coffe
        self.suger_coffe = suger_coffe
        self.prim_coffe = prim_coffe 
        self.milk_coffe = milk_coffe
        self.money = money
        
    # 돈받기
    def inmoney(self, cash):
        self.money += cash
        
    # 잔액반환   
    def outmoney(self, cash):
        if money > self.cash:
            return False
          
        else:
            self.cash -= money
            return money
       
```

```python
def show_menu():
    print('================================')
    print('======  커피 자동 판매기 동작  ======')
    print('================================')
    print("1. 블랙커피:(",cost_change['블랙커피'],')원')
    print("2. 설탕커피:(",cost_change['설탕커피'],')원')
    print("3. 프림커피:(",cost_change['프림커피'],')원')
    print("4. 밀크커피:(",cost_change['밀크커피'],')원')
    print("5. 종료")
    print('--------------------------------')
    
def manager_menu():
    print('==== 자동판매기 관리 프로그램 ====')
    print('[1]재료보충 [2]금액조정 [3]작업종료')
    print('============================')   
    

def refill():
    print('\t===== 재료보충 =====\t')       
    print('[1]전체 [2]커피 [3]물 [4]설탕 [5]프림 [6]컵')       
    print('======================================')  
    print('*메뉴에 없는 번호를 누르면 작업이 종료됩니다.*')
   
    
def modify():
    print('\t===== 금액조정 =====\t')
    print('금액 조정할 메뉴를 선택해주세요.')
    print('[1]블랙커피 [2]설탕커피 [3]프림커피 [4]밀크커피 [5]작업종료')       
    print('======================================')    
    




def manager_pg():
    while True:
        manager_menu()
        select_num = int(input('메뉴 = '))

        if select_num == 1:
            refill_pg()
        elif select_num == 2:
            modify_pg()
        elif select_num == 3:
            break
        else:
            print("메뉴의 선택이 올바르지 않습니다.")
    
def refill_pg():
     while True:
        refill()
        check_num = int(input('선택 = '))

        if check_num == 1:
            material['coffe'] +3
            material['water'] +3
            material['suger'] +3
            material['prim'] +3
            material['cup'] +3
            print('전체 추가가 완료되었습니다.')
        elif check_num == 2:
            material['coffe'] +3
            print('커피 추가가 완료되었습니다.')

        elif check_num == 3:
            material['water'] +3
            print('물 추가가 완료되었습니다.')

        elif check_num == 4:
            material['suger'] +3
            print('설탕 추가가 완료되었습니다.')
        elif check_num == 5:
            material['prim'] +3
            print('프림 추가가 완료되었습니다.')
        elif check_num == 6:
            material['cup'] +3
            print('컵 추가가 완료되었습니다.')
        else:
            print("메뉴의 선택이 올바르지 않습니다.")
            print("작업을 종료합니다.") 
            break# 만약 안되면 각 숫자 다음에 menu()집어넣기
       
    
def modify_pg():
        modify()
        modify_num = int(input('선택 = '))
        change = int(input('변경하실 금액을 입력하세요.'))

        if modify_num == 1:
            cost_change['블랙커피'] = change
        elif modify_num == 2:
            cost_change['설탕커피'] = change
        elif modify_num == 3:
            cost_change['프림커피'] = change
        elif modify_num == 4:
            cost_change['밀크커피'] = change
        elif modify_num == 5:
            break
            show_menu()
        else:
            print("메뉴의 선택이 올바르지 않습니다.")

        print("작업을 종료합니다.")
```
